```python
availability = []
for i in collection.find( {}, {'_id':1, 'availability':1}):
    availability.append(i)

df_availability = pd.DataFrame(availability)
availability_keys = list(df_availability.iloc[0,1].keys())

for i in availability_keys:
    df_availability['availability_30'] = df_availability['availability'].apply(lambda x: x['availability_30'])
    df_availability['availability_60'] = df_availability['availability'].apply(lambda x: x['availability_60'])
    df_availability['availability_90'] = df_availability['availability'].apply(lambda x: x['availability_90'])
    df_availability['availability_365'] = df_availability['availability'].apply(lambda x: x['availability_365'])

df_availability.drop(columns=['availability'], inplace=True)
df_availability.head()
```

```python
df_airbnb = pd.DataFrame(Airbnb_data)

# Merge df_airbnb with df_availability on a common key if available
df = pd.merge(df_airbnb, df_availability, on='_id', how='inner')
df.head()
```

```python
# Convert columns to relevant data types

# Convert 'Price' column to float
df['Price'] = df['Price'].astype(str).astype(float)

# Convert 'Security_deposit' column to float, handling NaN values
df['Security_deposit'] = df['Security_deposit'][~df['Security_deposit'].isna()].astype(str).astype(float)

# Convert 'Cleaning_fee' column to float, handling NaN values
df['Cleaning_fee'] = df['Cleaning_fee'][~df['Cleaning_fee'].isna()].astype(str).astype(float)

# Convert 'Extra_people' column to float
df['Extra_people'] = df['Extra_people'].astype(str).astype(float)

# Convert 'Guests_included' column to float
df['Guests_included'] = df['Guests_included'].astype(str).astype(float)

# Convert 'Review_scores' column to nullable integers
df['Review_scores'] = df['Review_scores'].astype('Int64')
```

```python
df.isna().sum()
```

```python
# Fill missing values in 'Total_bedrooms' with the mode
df['Total_bedrooms'].fillna(df['Total_bedrooms'].mode()[0], inplace=True)

# Fill missing values in 'Total_beds' with the median due to outliers
df['Total_beds'].fillna(df['Total_beds'].median(), inplace=True)

# Fill missing values in 'Security_deposit' with the median
df['Security_deposit'].fillna(df['Security_deposit'].median(), inplace=True)

# Fill missing values in 'Cleaning_fee' with the median
df['Cleaning_fee'].fillna(df['Cleaning_fee'].median(), inplace=True)

# Fill missing values in 'Review_scores' with the median
df['Review_scores'].fillna(df['Review_scores'].median(), inplace=True)
```

```python
# Filling Empty values in Description and House rules columns
df.Description.replace(to_replace='',value='No Description Provided',inplace=True)
df.House_rules.replace(to_replace='',value='No House rules Provided',inplace=True)
df.Amenities.replace(to_replace='',value='Not Available',inplace=True)
```

```python
df.isna().sum()
```

```python
# Drop rows with empty values in the 'Name' column
df.dropna(subset=['Name'], inplace=True)

# Drop rows with duplicate values in the 'Name' column
duplicate_name_indices = df[df['Name'].duplicated(keep=False)].index
df.drop(labels=duplicate_name_indices, inplace=True)
```

```python
# Reset the index of the DataFrame
df.reset_index(drop=True, inplace=True)
```
